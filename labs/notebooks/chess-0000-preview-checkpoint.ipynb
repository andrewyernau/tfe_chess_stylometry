```python
# Importación de librerías
%pip install python-chess
import chess.pgn
from IPython.display import display, SVG
from pathlib import Path
import os

# Obtener el archivo PGN
DATA_DIRECTORY : Path = Path("..")/"dataset";
pgn_path = DATA_DIRECTORY / "chessgame0001.pgn"
START_MOVE = 15
END_MOVE = 23
with open(pgn_path, encoding="utf-8") as pgn:
    game = chess.pgn.read_game(pgn)

# Crear tablero
board = game.board()

# Especificar hasta qué movimiento queremos mostrar por ejemplo 15
target_move = START_MOVE

# Reproducir los movimientos hasta el movimiento objetivo
moves = list(game.mainline_moves())
for i, move in enumerate(moves):
    if i >= target_move:
        break
    board.push(move)

print(f"Mostrando la posición después del movimiento {target_move}")
print(f"Total de movimientos en la partida: {len(moves)}")
print(type(game))
# Mostrar el tablero como SVG en el notebook
display(SVG(board._repr_svg_()))
```

```python
# Movimientos 15 hasta la 31,
start_move = START_MOVE
end_move = END_MOVE

# Obtener todos los movimientos de la partida
all_moves = list(game.mainline_moves())

# Extraer el rango de movimientos deseado (ajustando índices porque empiezan en 0)
selected_moves = all_moves[start_move-1:end_move-1]

print(f"Movimientos desde la jugada {start_move} hasta la {end_move}:")
print(f"Total de movimientos extraídos: {len(selected_moves)}")
print()

# Mostrar los movimientos en formato legible
for i, move in enumerate(selected_moves, start=start_move):
    print(f"Jugada {i}: {move}")

# Almacenar en array para uso posterior
moves_array = selected_moves
print(f"\nArray de movimientos almacenado con {len(moves_array)} elementos")
```

```python
%pip install numpy cairosvg pandas
import sys
import numpy as np
import cairosvg
from PIL import Image
import io
import pandas as pd

# Convertir SVG a PNG usando cairosvg
def svg_to_array(svg_string, size=(224, 224)):
    png_data = cairosvg.svg2png(bytestring=svg_string.encode('utf-8'))
    
    # Convertir PNG a PIL Image
    image = Image.open(io.BytesIO(png_data))
    
    # Redimensionar si es necesario
    if image.size != size:
        image = image.resize(size, Image.Resampling.LANCZOS)
    
    # Convertir a RGB si no lo está
    if image.mode != 'RGB':
        image = image.convert('RGB')
    
    # Convertir a array numpy y normalizar
    image_array = np.array(image) / 255.0
    
    return image_array

# Generar posiciones e imágenes desde jugada 15 hasta 31
start_move = START_MOVE
end_move = END_MOVE
board_positions = []
board_images = []
move_numbers = []

# Crear un nuevo tablero para recorrer las posiciones
board_temp = game.board()
all_moves = list(game.mainline_moves())

# Avanzar hasta la posición inicial (jugada START_MOVE)
for i in range(start_move):
    if i < len(all_moves):
        board_temp.push(all_moves[i])

# Capturar la posición inicial (jugada START_MOVE)
svg_string = board_temp._repr_svg_()
image_array = svg_to_array(svg_string)
board_positions.append(board_temp.fen())
board_images.append(image_array)
move_numbers.append(start_move)

print(f"Capturada posición {start_move}")

# Capturar posiciones siguientes hasta jugada END_MOVE
for move_num in range(start_move + 1, min(end_move + 1, len(all_moves) + 1)):
    if move_num - 1 < len(all_moves):
        board_temp.push(all_moves[move_num - 1])
        
        svg_string = board_temp._repr_svg_()
        image_array = svg_to_array(svg_string)
        
        board_positions.append(board_temp.fen())
        board_images.append(image_array)
        move_numbers.append(move_num)
        
        print(f"Capturada posición {move_num}")

# Crear DataFrame con las posiciones e imágenes
chess_data = pd.DataFrame({
    'move_number': move_numbers,
    'fen_position': board_positions,
    'board_image': board_images
})

# Convertir imágenes a array numpy 4D para CNN (batch_size, height, width, channels)
images_array = np.stack(board_images)

print(f"\nDatos generados:")
print(f"- Total de posiciones: {len(board_positions)}")
print(f"- Shape del array de imágenes: {images_array.shape}")
print(f"- Tamaño del DataFrame: {chess_data.shape}")
print(f"- Rango de movimientos: {min(move_numbers)} - {max(move_numbers)}")

chess_data.head()
```

```python
# Previsualizar las imágenes PNG generadas
%pip install matplotlib
import matplotlib.pyplot as plt

def show_chess_positions(chess_data, num_images=END_MOVE-START_MOVE+1):
    fig, axes = plt.subplots(3, 3, figsize=(15, 10))
    axes = axes.flatten()
    
    # Seleccionar imágenes distribuidas uniformemente
    total_images = len(chess_data)
    indices = np.linspace(0, total_images-1, num_images, dtype=int)
    
    for i, idx in enumerate(indices):
        if i < len(axes):
            # Obtener la imagen y la información
            image = chess_data.iloc[idx]['board_image']
            move_num = chess_data.iloc[idx]['move_number']
            
            # Mostrar la imagen
            axes[i].imshow(image)
            axes[i].set_title(f'Posición después del movimiento {move_num}', fontsize=12)
            axes[i].axis('off')
    
    plt.tight_layout()
    plt.show()

# Mostrar información sobre las imágenes
print("Información sobre las imágenes generadas:")
print(f"- Formato: {chess_data.iloc[0]['board_image'].shape}")
print(f"- Tipo de datos: {chess_data.iloc[0]['board_image'].dtype}")
print(f"- Rango de valores: [{chess_data.iloc[0]['board_image'].min():.3f}, {chess_data.iloc[0]['board_image'].max():.3f}]")
print()

# Mostrar una selección de las posiciones
show_chess_positions(chess_data, num_images=9)

# Mostrar también la primera imagen individual para verificar calidad
print(f"\nPrimera imagen individual (posición {START_MOVE}):")
plt.figure(figsize=(6, 6))
plt.imshow(chess_data.iloc[0]['board_image'])
plt.title(f"Posición después del movimiento {chess_data.iloc[0]['move_number']}")
plt.axis('off')
plt.show()
```

```python
# Mostrar una imagen específica por número de jugada
def show_specific_position(chess_data, target_move):
    # Buscar la fila correspondiente al movimiento objetivo
    target_row = chess_data[chess_data['move_number'] == target_move]
    
    if target_row.empty:
        available_moves = sorted(chess_data['move_number'].tolist())
        print(f"Movimiento {target_move} no encontrado.")
        print(f"Movimientos disponibles: {available_moves}")
        return
    
    # Obtener los datos de la posición
    image = target_row.iloc[0]['board_image']
    fen_position = target_row.iloc[0]['fen_position']
    
    # Mostrar la imagen
    plt.figure(figsize=(8, 8))
    plt.imshow(image)
    plt.title(f'Posición después del movimiento {target_move}', fontsize=16, fontweight='bold')
    plt.axis('off')
    
    # Agregar información adicional
    plt.figtext(0.1, 0.02, f'FEN: {fen_position}', fontsize=10, wrap=True)
    plt.tight_layout()
    plt.show()
    
    print(f" Mostrando posición después del movimiento {target_move}")
    print(f" FEN: {fen_position}")

# Mostrar la posición END_MOVE
target_position = END_MOVE
print(f"Buscando y mostrando la posición {target_position}:")
show_specific_position(chess_data, target_position)

print(f"\n{'='*50}")
print("Para ver otra posición, cambia el valor de 'target_position' y ejecuta de nuevo.")
print(f"Posiciones disponibles: {sorted(chess_data['move_number'].tolist())}")
```

```python
# Convertir todas las imágenes a escala de grises
def rgb_to_grayscale(image_array):
    # Pesos estándar para conversión RGB a escala de grises
    # R: 0.299, G: 0.587, B: 0.114
    grayscale = np.dot(image_array[...,:3], [0.299, 0.587, 0.114])
    return grayscale

print("Convirtiendo imágenes a escala de grises...")
grayscale_images = []

for i, image in enumerate(chess_data['board_image']):
    gray_image = rgb_to_grayscale(image)
    grayscale_images.append(gray_image)
    
    if (i + 1) % 5 == 0 or i == len(chess_data) - 1:
        print(f"Procesadas {i + 1}/{len(chess_data)} imágenes")

# Crear nuevo DataFrame con imágenes en escala de grises
chess_data_gray = chess_data.copy()
chess_data_gray['board_image_gray'] = grayscale_images

# Convertir a array numpy
images_gray_array = np.stack(grayscale_images)
images_gray_array = np.expand_dims(images_gray_array, axis=-1)  # Agregar dimensión de canal

print(f"  Información de las imágenes en escala de grises:")
print(f"- Shape original (RGB): {chess_data.iloc[0]['board_image'].shape}")
print(f"- Shape escala de grises: {grayscale_images[0].shape}")
print(f"- Shape array 4D para CNN: {images_gray_array.shape}")
print(f"- Rango de valores: [{images_gray_array.min():.3f}, {images_gray_array.max():.3f}]")

# RGB vs Escala de grises
def compare_rgb_vs_grayscale(chess_data_gray, move_number):
    target_row = chess_data_gray[chess_data_gray['move_number'] == move_number]
    
    if target_row.empty:
        print(f"Movimiento {move_number} no encontrado")
        return
    
    rgb_image = target_row.iloc[0]['board_image']
    gray_image = target_row.iloc[0]['board_image_gray']
    
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))
    
    # Imagen RGB original
    ax1.imshow(rgb_image)
    ax1.set_title(f'RGB - Movimiento {move_number}', fontsize=14)
    ax1.axis('off')
    
    # Imagen en escala de grises
    ax2.imshow(gray_image, cmap='gray')
    ax2.set_title(f'Escala de Grises - Movimiento {move_number}', fontsize=14)
    ax2.axis('off')
    
    plt.tight_layout()
    plt.show()

# Ejemplo
print(f"\n Comparación RGB vs Escala de Grises para el movimiento {END_MOVE}:")
compare_rgb_vs_grayscale(chess_data_gray, END_MOVE)

chess_data_gray.head()
```

```python
# Filtramos las imágenes del rango deseado
mask = (chess_data_gray['move_number'] >= START_MOVE) & (chess_data_gray['move_number'] <= END_MOVE)
sequence = np.stack(chess_data_gray.loc[mask, 'board_image_gray'].values)

# sequence.shape = (num_frames, H, W)
print(f"Secuencia seleccionada: {sequence.shape}")

# Normalizamos a [0,1]
sequence = sequence.astype(np.float32)
sequence /= sequence.max() if sequence.max() > 1 else 1.0

# Si quieres un tensor CNN-friendly: (H, W, num_frames)
stacked_board = np.moveaxis(sequence, 0, -1)
print(f"Stacked board shape: {stacked_board.shape}")
```

```python
# Dividimos las 9 jugadas en 3 grupos de 3 y las comprimimos
channels = []
for i in range(0, 9, 3):
    chunk = sequence[i:i+3].mean(axis=0)
    channels.append(chunk)

# Creamos imagen RGB
cnn_friendly_image = np.stack(channels, axis=-1)
cnn_friendly_image = (cnn_friendly_image * 255).astype(np.uint8)

print(f"CNN-friendly image shape: {cnn_friendly_image.shape}")

import cv2
cv2.imwrite("cnn_friendly_board_15_23.png", cnn_friendly_image)
```
